```python
def date_format(s):
    
    s_list = s.split("/")
    
    year = s_list[0]
    month = int(s_list[1])
    day = int(s_list[2])
    
    if(month<10):
        month = "0"+str(month)
    else:
        month = str(month)
    
    if(day<10):
        day = "0"+str(day)
    else:
        day = str(day)

    s0 = year+month+day
    
    return s0
```

```python
hs300_dict = {}

for i in range(2000):
    
    date = date_format(hs300['日期'][i])
    return_rate = float(hs300['涨跌幅(%)'][i])
    
    hs300_dict[date] = return_rate
```

```python
import math

def calculate_ab_return(stock):
    
    stock_name = stock.columns.values.tolist()
    stock_name = stock_name[1:-3]
    
    for i in range(len(stock)):
        
        date = str(stock.trade_date[i])
        index_rate = hs300_dict[date]
        
        all_stock = 0
        positive_stock = 0
        abr_list = []
        
        for s in stock_name:
            
            if(math.isnan(stock[s][i])==False):
                
                all_stock = all_stock+1
                
                stock[s][i] = float(stock[s][i]) - index_rate
                abr_list.append(stock[s][i])
                
                if(stock[s][i]>0):
                    positive_stock = positive_stock+1
        
        stock["positive"][i] = float(positive_stock/all_stock)
        stock["max_abr"][i] = float(max(abr_list))
        stock["min_abr"][i] = float(min(abr_list))

    return stock

stock_ab_return = calculate_ab_return(stock)
stock_ab_return
```

```python
# 累计超额收益率 

def accumulate_abnormal_return(df,abnormal,day):
    
#     date_list = list(df['trade_date'])

    df = df.rename({'max_abr': 'max_acr', 'min_abr': 'min_acr'}, axis=1)
    df['positive'] = 0.0
    df['max_acr'] = 0.0
    df['min_acr'] = 0.0
    
    stock_name = df.columns.values.tolist()
    stock_name = stock_name[1:-3]
    
    for i in range(len(df)):
        
        if((i+day)<=len(df)):
            
            all_stock = 0
            positive_stock = 0
            acr_list = []
        
            for stock in stock_name:
                 
                if(math.isnan(abnormal[stock][i+day-1])==True):
                    acr = math.nan
                    
                else:
                    
                    all_stock = all_stock+1
                    
                    acr = 0
                    for j in range(day):
                        acr = acr+ float(abnormal[stock][i+j])
                    
                    acr_list.append(acr)
                    if(acr>0):
                        positive_stock = positive_stock+1
                        

                df[stock][i] = acr
            
            df['positive'][i] = positive_stock/all_stock
            df['max_acr'][i] = max(acr_list)
            df['min_acr'][i] = min(acr_list)
    
    return df
```

```python
df = pd.read_csv("data/price/abnormal_return.csv")
abnormal = pd.read_csv("data/price/abnormal_return.csv")

df3 = accumulate_abnormal_return(df,abnormal,3)
```

```python
# 汇总已有的情感词典

import pandas as pd
pd.options.mode.chained_assignment = None  # default='warn'

# 中文情感词典 姜富伟等
# 知网
# NTUSD
# 清华李军
# DLUT
# cheatsheet
```

```python
positive = pd.read_csv("../已有情感词典/positive_combine.csv")
positive
```

```python
from langconv import *

# 将繁体转换成简体
    
def tradition2simple(line):
    
    line = Converter('zh-hans').convert(line)
#     line = line.encode('utf-8')
    
    return line
```

```python
def remove_dup(df):
    
    word_list = list(df['word'])
    word_list = list(set(word_list))
    df = pd.DataFrame(word_list, columns = ['word'])
    
    return df
```

```python
for i in range(len(positive)):
    
    positive.word[i] = positive.word[i].replace("\n", "")
    positive.word[i] = tradition2simple(positive.word[i])
```

```python
positive
```

```python
positive = remove_dup(positive)
positive
```

```python
positive.to_csv("../已有情感词典/positive_combine.csv",index=False, encoding='utf_8_sig')
```

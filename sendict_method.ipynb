```python
import pandas as pd
pd.options.mode.chained_assignment = None  # default='warn'
```

```python
# 返回 dict

def rate_for_report(words,code,date,accr3,accr5,accr3_label,accr5_label,positive,negative):
    
    positive_num = 0
    negative_num = 0
    
    for word in words:
        
        if(word in positive):
            positive_num = positive_num+1
            
        if(word in negative):
            negative_num = negative_num+1
    
    one_dict = {}
    one_dict['code'] = code
    one_dict['date'] = date
    one_dict['accr3'] = accr3
    one_dict['accr3_label'] = accr3_label
    one_dict['accr5'] = accr5
    one_dict['accr5_label'] = accr5_label
    one_dict['positive'] = positive_num
    one_dict['negative'] = negative_num
    one_dict['length'] = len(words)
    
    return one_dict
```

```python
def rate_dateframe(df,stock_code,positive,negative):
    
    df = df.dropna()
    df = df.reset_index(drop=True)
    
    data_list = []
    
    for i in range(len(df)):
        
        words = list(df.words[i].split("/"))
        code = stock_code
        date = df.ann_date[i]
        accr3 = df.accr3[i]
        accr3_label = df.accr3_label[i]
        accr5 = df.accr5[i]
        accr5_label = df.accr5_label[i]
        
        one_dict = rate_for_report(words,code,date,accr3,accr5,accr3_label,accr5_label,positive,negative)
        
        data_list.append(one_dict)
    
    return data_list
```

```python
def sentiment_dict_rate(filename,t1,t2):
    
    positive = pd.read_csv("../已有情感词典/positive_combine.csv")
    positive = list(positive['word'])

    negative = pd.read_csv("../已有情感词典/negative_combine.csv")
    negative = list(negative['word'])
    
    df_hs300 = pd.read_csv('data/HS300/' + filename + '.csv')
    stock_code = list(df_hs300['成份代码'])
    foldername = t1+'-'+t2
    
    result = []

    for i in range(len(stock_code)):
        
        one_stock_code = stock_code[i]
        savefilepath = 'data/HS300/' + foldername + '/'+ one_stock_code+'.csv'
        df_report = pd.read_csv(savefilepath)
        
        result0 = rate_dateframe(df_report,one_stock_code,positive,negative)
        result.extend(result0)
        
        if(i%15==0):
            print(i)
    
    result = pd.DataFrame(result)
    
    return result
#     result.to_csv('data/HS300/' + foldername + '/'+ foldername +'_dict_rate.csv')
```

```python
# 2016 1

sentiment_dict_rate('2015-12-30','20151230','20160613')
```

```python
# 2017 1

result = sentiment_dict_rate('2016-12-12','20161212','20170612')
```

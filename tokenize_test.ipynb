```python
# 去空格

df['content_clean'] = ''
df['content_clean'][0] = df['content'][0].replace(' ','')
df['content_clean'][0]
```

```python
#去除字母数字表情和其它字符

import re

def clear_character(sentence):
    
    pattern1='[a-zA-Z0-9]'
#     pattern1='[0-9]'
    pattern2 = '\[.*?\]'
    pattern3 = re.compile(u'[^\s1234567890:：' + '\u4e00-\u9fa5]+')
    pattern4='[’!"#$%&\'()*+,-./：:;<=>?@[\\]^_`{|}~]+'
    
    line1=re.sub(pattern1,'',sentence)   #去除英文字母和数字
#     line1 = sentence
    line1=re.sub(pattern1,'',sentence)   #去除数字
    line2=re.sub(pattern2,'',line1)      #去除表情
    line3=re.sub(pattern3,'',line2)      #去除其它字符
#     line3=line2
    line4=re.sub(pattern4, '', line3)    #去掉残留的冒号及其它符号
    new_sentence=''.join(line4.split())  #去除空白
    
    return new_sentence
```

```python
df['content_clean'][0] = clear_character(df['content'][0])
df['content_clean'][0]
```

```python
import jieba
import paddle
paddle.enable_static()

jieba.load_userdict("data/dictionary/user_dict.txt")  # 加入自定义金融词典
jieba.enable_paddle()# 启动paddle模式
```

```python
seg_list = jieba.cut(df['content_clean'][0])  # 精确模式
print(", ".join(seg_list))
```

```python
seg_list = jieba.cut(df['content_clean'][0],use_paddle=True)  # 使用paddle模式
print(", ".join(seg_list))
```

```python
# 加入去停词

def stopwordslist(filepath):  
    
    stopwords = [line.strip() for line in open(filepath, 'r', encoding='utf-8').readlines()]  
    return stopwords  
 

def seg_sentence(sentence,use_paddle=False):  
    
    sentence_seged = jieba.cut(sentence, use_paddle=use_paddle)  
    stopwords = stopwordslist('data/stopwords/user_stopwords.txt')  # 这里加载停用词的路径  
    outstr = ''  
    for word in sentence_seged:  
        if word not in stopwords:  
            if word != '\t':  
                outstr += word  
                outstr += " "   
    return outstr  

seg = seg_sentence(df['content_clean'][0])    # 选择精确模式
seg
```

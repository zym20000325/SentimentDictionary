```python
def count_word(df,report,accr3,accr5):    
    
    # report is a tokenized word list
    
    word_list = list(df['word'])
    
    for word in word_list:
        
        if(word in report):
            
            index = df[df['word']==word].index.values[0]
            
            df['occurrence'][index] = df['occurrence'][index]+1
            df['ave_accr3'][index] = df['ave_accr3'][index]+accr3
            df['ave_accr5'][index] = df['ave_accr5'][index]+accr5
            
            if(accr3>0):
                df['positive3'][index] = df['positive3'][index]+1
            
            if(accr5>0):
                df['positive5'][index] = df['positive5'][index]+1
        
    
    return df
```

```python
def browse_report(df,df_report):
    
    df_report = df_report.dropna()
    df_report = df_report.reset_index(drop=True)
    
    for i in range(len(df_report)):
        
        report = list(df_report.words[i].split("/"))
        accr3 = float(df_report.accr3[i])
        accr5 = float(df_report.accr5[i])
        
        df = count_word(df,report,accr3,accr5)
    
    return df
```

```python
def count_corpus_word(df,filename,t1,t2):
    
    df_hs300 = pd.read_csv('data/HS300/' + filename + '.csv')
    stock_code = list(df_hs300['成份代码'])
    foldername = t1+'-'+t2

    for i in range(len(stock_code)):
        
        one_stock_code = stock_code[i]
        savefilepath = 'data/HS300/' + foldername + '/'+ one_stock_code+'.csv'
        df_report = pd.read_csv(savefilepath)
        
        df = browse_report(df,df_report)
        
        if(i%15==0):
            print(i)
    
    return df
```

```python
def initialize_df(path):
    
    df = pd.read_csv(path)
    
    df['occurrence'] = 0
    df['positive3'] = 0
    df['ave_accr3'] = 0.0
    df['positive5'] = 0
    df['ave_accr5'] = 0.0
    
    return df
```

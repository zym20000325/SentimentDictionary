```python
corpus = pd.read_csv('data/HS300/corpus_all8.csv')
corpus = list(corpus['word'])
len(corpus)
```

```python
cover_num = 0
count = 0
words_covered = []

with open('../sgns.financial.word',  encoding='utf_8_sig', errors='ignore') as f:   # 真是天才
    for line in f:
        
        sample = line
        sample = list(sample.split(" "))
        dict_word = sample[0]
        if(dict_word in corpus):
            words_covered.append(dict_word)
            cover_num = cover_num+1

cover_num
```

```python
def diff_list(list1,list2):
    
    result = list(set(list1).difference(set(list2)))   # list1中有 而list2中没有
    return result
        
```
